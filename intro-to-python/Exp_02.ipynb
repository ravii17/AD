```python
#Plot heatmap of correlation matrix
import pandas as pd

# Create a temporary DataFrame to clean for correlation calculation
df4_cleaned = df4.copy()

# Replace problematic string values with NaN in the relevant columns
df4_cleaned['SepalLengthCm'] = df4_cleaned['SepalLengthCm'].replace('??', pd.NA)
df4_cleaned['PetalLengthCm'] = df4_cleaned['PetalLengthCm'].replace('###', pd.NA)

# Convert the specified columns to numeric, coercing any remaining errors to NaN
df4_cleaned['SepalLengthCm'] = pd.to_numeric(df4_cleaned['SepalLengthCm'], errors='coerce')
df4_cleaned['SepalWidthCm'] = pd.to_numeric(df4_cleaned['SepalWidthCm'], errors='coerce')
df4_cleaned['PetalLengthCm'] = pd.to_numeric(df4_cleaned['PetalLengthCm'], errors='coerce')
df4_cleaned['PetalWidthCm'] = pd.to_numeric(df4_cleaned['PetalWidthCm'], errors='coerce')

# Drop rows that have NaN values in any of the columns used for correlation calculation
df4_cleaned = df4_cleaned.dropna(subset=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'])

sns.heatmap(data=df4_cleaned[['SepalLengthCm','SepalWidthCm',
'PetalLengthCm','PetalWidthCm']].corr(),
annot=True, linecolor='white', linewidths=0.25, cbar=True,
            cmap='viridis')

plt.title('Correlation Matrix of Iris Features')
plt.show()
```

```python
#Label Encoder
from sklearn.preprocessing import LabelEncoder
data = ['red', 'blue', 'green', 'blue', 'red']
label_encoder = LabelEncoder()
encoded_data = label_encoder.fit_transform(data)
print(encoded_data)
```

```python
#OneHot Encoder
from sklearn.preprocessing import OneHotEncoder
import numpy as np
# Sample data
data = np.array(['red', 'blue', 'green', 'blue', 'red']).reshape(-1, 1)
# Initialize OneHotEncoder
onehot_encoder = OneHotEncoder(sparse_output=False)
# Fit and transform the data
encoded_data = onehot_encoder.fit_transform(data)
print(encoded_data)
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np
data = np.array(['red', 'blue', 'green', 'blue', 'red']).reshape(-1, 1)
onehot_encoder = OneHotEncoder(sparse_output=False)
encoded_data = onehot_encoder.fit_transform(data)
print(encoded_data)
```

```python
from sklearn.preprocessing import OrdinalEncoder
data = np.array(['low', 'medium', 'high', 'medium', 'low']).reshape(-1, 1)
ordinal_encoder = OrdinalEncoder(categories=[['low', 'medium', 'high']])
encoded_data = ordinal_encoder.fit_transform(data)
print(encoded_data)
```
